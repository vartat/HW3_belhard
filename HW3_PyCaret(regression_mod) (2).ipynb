```python
from pycaret.datasets import get_data
data = get_data('gold')
#[column for column in data]
```

```python
import pandas as pd
corrs=data.corr()
corrs.style.background_gradient(cmap='coolwarm')
```

```python
#выводим 25 пар признаков с максимальной корреляцией
corrs = [(corrs.iloc[row,col], row, col) for row in range(1, 121) for col in range(row)]
max_cor = sorted(corrs, key=lambda x: abs(x[0]), reverse=True)[:25]
print(max_cor)
```

```python
#удаляем столбцы с корреляцией по модулю выше 0,95
cols = [11, 41, 71, 101, 67, 97, 37, 94, 64, 7, 34, 115, 114, 84, 19, 107, 54, 49, 77, 79, 109, 85, 47, 90]
data.drop (data.columns [cols], axis= 1 , inplace= True)
```

```python
from pycaret.regression import *
s = setup(data, target = 'Gold_T+22', session_id = 123)
```

```python
best = compare_models()
```

```python
evaluate_model(best)
```

```python
predictions = predict_model(best, data=data)
```

```python
save_model(best, 'gold_regression_modif')
```
